```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1=pd.read_excel("BREIT - Histórico de notas Champs.xlsx",sheet_name='Promedio de notas 2023')
df1['Year']=2023
df2=pd.read_excel("BREIT - Histórico de notas Champs.xlsx",sheet_name='Promedio de notas 2022')
df2['Year']=2022
df3=pd.read_excel("BREIT - Histórico de notas Champs.xlsx",sheet_name='Promedio de notas 2021')
df3['Year']=2021
df4=pd.read_excel("BREIT - Histórico de notas Champs.xlsx",sheet_name='Promedio de notas 2020')
df4['Year']=2020
df5=pd.read_excel("BREIT - Histórico de notas Champs.xlsx",sheet_name='Promedio de notas 2019')
df5['Year']=2019
df6=pd.read_excel("BREIT - Histórico de notas Champs.xlsx",sheet_name='Promedio de notas 2018')
df6['Year']=2018
```

```python
df=pd.concat([df1,df2,df3,df4,df5,df6])
```

```python
# Supongamos que df es tu DataFrame
# Primero, calculamos la mediana de 'Conducta' usando datos de años anteriores
mediana_conducta = df[df['Year'] != 2020]['Conducta'].median()

# Ahora imputamos la mediana en los valores faltantes de 'Conducta' para el año 2020
df.loc[df['Year'] == 2020, 'Conducta'] = df.loc[df['Year'] == 2020, 'Conducta'].fillna(mediana_conducta)
```

```python
# Crear una figura con múltiples subplots
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(10, 20))  # Ajusta según el número de variables
axes = axes.flatten()  # Aplanar el array de ejes si es multidimensional

subjects = ['Matemática', 'Comunicación', 'Ciencias', 'Dpsc', 'Educación Física', 'Arte', 'Conducta', 'PROMEDIO', 'PREFINAL', 'GRADO','CANTERA']

# Iterar a través de los subjects y graficar cada histograma en su propio eje
for idx, subject in enumerate(subjects):
    # Asegúrate de que los datos son numéricos, especialmente para 'GRADO' que es categórico
    if df[subject].dtype == 'O':  # Si el tipo de objeto es 'O' (objeto), no intentar histograma
        df[subject].value_counts().plot(kind='bar', ax=axes[idx])
    else:
        df[subject].hist(bins=5, ax=axes[idx], alpha=0.7)
    axes[idx].set_title(f'Histograma de {subject}')
    axes[idx].set_xlabel('Valor')
    axes[idx].set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

```python
# Crear una figura con múltiples subplots
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(10, 20))
axes = axes.flatten()  # Aplanar el array de ejes si es multidimensional

subjects = ['Matemática', 'Comunicación', 'Ciencias', 'Dpsc', 'Educación Física', 'Arte', 'Conducta', 'PROMEDIO', 'PREFINAL']

# Mapeo de colores para 'GRADO'
color_mapping = {
    'Octavo': 'red',
    'Séptimo': 'blue',
    'Undécimo': 'green',
    'Décimo': 'purple'
}
df['color'] = df['GRADO'].map(color_mapping)

# Iterar a través de los subjects y graficar cada histograma en su propio eje
for idx, subject in enumerate(subjects):
    if df[subject].dtype == 'O':  # Si el tipo de objeto es 'O' (objeto), no intentar histograma
        continue
    # Crear histograma con colores basados en 'GRADO'
    for grade, color in color_mapping.items():
        subset = df[df['GRADO'] == grade]
        axes[idx].hist(subset[subject], bins=5, alpha=0.7, label=grade, color=color)

    axes[idx].set_title(f'Histograma de {subject}')
    axes[idx].set_xlabel('Valor')
    axes[idx].set_ylabel('Frecuencia')
    axes[idx].legend(title='Grado')

plt.tight_layout()
plt.show()
```

```python
# Mapeo de colores para 'CANTERA'
canteras = df['CANTERA'].unique()
colors = plt.cm.viridis(np.linspace(0, 1, len(canteras)))
color_mapping = dict(zip(canteras, colors))

# Crear una figura para cada categoría en 'CANTERA'
for cantera in canteras:
    fig, axes = plt.subplots(nrows=len(subjects), ncols=1, figsize=(10, 30))
    fig.suptitle(f'Histogramas para la cantera: {cantera}', fontsize=16)
    
    for idx, subject in enumerate(subjects):
        if df[subject].dtype == 'O':  # Evitar intentar histograma de datos no numéricos
            continue
        
        subset = df[df['CANTERA'] == cantera]
        axes[idx].hist(subset[subject], bins=25, color=color_mapping[cantera], alpha=0.7)
        axes[idx].set_title(subject)
        axes[idx].set_xlabel('Valor')
        axes[idx].set_ylabel('Frecuencia')
    
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Ajustar el layout para incluir el título principal
    plt.show()
```
